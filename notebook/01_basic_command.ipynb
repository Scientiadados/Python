# 5️⃣ Criando colunas novas (transformação)

```python
df["total"] = df["quantidade"] * df["preco"]
df.head()
```

```python
import datetime 
df['data'] = pd.to_datetime(df['data'], format='%d/%m/%Y')  # Para formato dd/mm/yyyy
df["ano"] = df["data"].dt.year
df["mes"] = df["data"].dt.month
df.head()
```

# 6️⃣ Alterando valores

```python
df["cidade"] = df["cidade"].str.upper()
df.head()
```

```python
df.loc[df["quantidade"] == 0, "quantidade"] = 1
df.head()
```
